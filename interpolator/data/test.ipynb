```python
import numpy as np
import matplotlib.pyplot as plt

nodes = np.loadtxt('nodes.txt')
x_node = nodes[:, 0] * 0.25
y_node = nodes[:, 1]

data = np.loadtxt('data.txt')
x = data[:, 0]
y = data[:, 1]

plt.figure()
plt.plot(x, y)
plt.plot(x_node, y_node, 'ro')
plt.grid()
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.show()

# plot the velocity using finite difference
plt.figure()
plt.plot(x[1:], np.diff(y) / np.diff(x))
plt.grid()
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.show()
```
