```python
import numpy as np
import matplotlib.pyplot as plt

# load data
t_og = np.loadtxt('t_original.txt')
data_og = np.loadtxt('data_original.txt')
t_array = np.loadtxt('t_array.txt')
data_array = np.loadtxt('data_array.txt')

# plot the data
plt.figure()
plt.plot(t_array, data_array[:, 0], 'r-')
plt.plot(t_og, data_og[:, 0], 'rx')
plt.plot(t_array, data_array[:, 1], 'b-')
plt.plot(t_og, data_og[:, 1], 'bx')
plt.plot(t_array, data_array[:, 2], 'g-')
plt.plot(t_og, data_og[:, 2], 'gx')
plt.xlabel('t')
plt.ylabel('data')
plt.grid()

# plot the second data
t_og2 = np.loadtxt('t_original2.txt')
data_og2 = np.loadtxt('data_original2.txt')
t_array2 = np.loadtxt('t_array2.txt')
data_array2 = np.loadtxt('data_array2.txt')

# plot the data
plt.figure()
plt.plot(t_array2, data_array2[:, 0], 'r-')
plt.plot(t_og2, data_og2[:, 0], 'rx')
plt.plot(t_array2, data_array2[:, 1], 'b-')
plt.plot(t_og2, data_og2[:, 1], 'bx')
plt.plot(t_array2, data_array2[:, 2], 'g-')
plt.plot(t_og2, data_og2[:, 2], 'gx')
plt.xlabel('t')
plt.ylabel('data')
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# import all the data
data_og = np.loadtxt('data_original.txt')
data_val = np.loadtxt('data.txt')
data_1deriv = np.loadtxt('data_derivative.txt')
data_2deriv = np.loadtxt('data_second_derivative.txt')

data_og2 = np.loadtxt('data_original2.txt')
data_val2 = np.loadtxt('data2.txt')
data_1deriv2 = np.loadtxt('data_derivative2.txt')
data_2deriv2 = np.loadtxt('data_second_derivative2.txt')

# plot the data
plt.figure(1)
plt.plot(data_val[:,0], data_val[:,1], 'b.')
plt.plot(data_val[:,0], data_val[:,2], 'r.')
plt.plot(data_og[:,0], data_og[:,1], 'bx', markersize=10)
plt.plot(data_og[:,0], data_og[:,2], 'rx', markersize=10)
plt.xlabel('t')
plt.ylabel('q ,v')
plt.title('Original data_og')
plt.legend(['q', 'v'])
plt.grid()

# plot derivatives
plt.figure(2)
plt.plot(data_1deriv[:,0], data_1deriv[:,1], 'b.')
plt.plot(data_1deriv[:,0], data_1deriv[:,2], 'r.')
plt.xlabel('t')
plt.ylabel('dq ,dv')
plt.title('First Derivative')
plt.legend(['dq', 'dv'])
plt.grid()

# plot the second derivatives
plt.figure(3)
plt.plot(data_2deriv[:,0], data_2deriv[:,1], 'b.')
plt.plot(data_2deriv[:,0], data_2deriv[:,2], 'r.')
plt.xlabel('t')
plt.ylabel('ddq ,ddv')
plt.title('Second Derivative')
plt.legend(['ddq', 'ddv'])
plt.grid()

# plot seoncd set of data
plt.figure(4)
plt.plot(data_val2[:,0], data_val2[:,1], 'b.')
plt.plot(data_val2[:,0], data_val2[:,2], 'r.')
plt.plot(data_og2[:,0], data_og2[:,1], 'bx', markersize=10)
plt.plot(data_og2[:,0], data_og2[:,2], 'rx', markersize=10)
plt.xlabel('t')
plt.ylabel('q ,v')
plt.title('Original data_og2')
plt.legend(['q', 'v'])
plt.grid()

# plot derivatives
plt.figure(5)
plt.plot(data_1deriv2[:,0], data_1deriv2[:,1], 'b.')
plt.plot(data_1deriv2[:,0], data_1deriv2[:,2], 'r.')
plt.xlabel('t')
plt.ylabel('dq ,dv')
plt.title('First Derivative')
plt.legend(['dq', 'dv'])
plt.grid()

# plot the second derivatives
plt.figure(6)
plt.plot(data_2deriv2[:,0], data_2deriv2[:,1], 'b.')
plt.plot(data_2deriv2[:,0], data_2deriv2[:,2], 'r.')
plt.xlabel('t')
plt.ylabel('ddq ,ddv')
plt.title('Second Derivative')
plt.legend(['ddq', 'ddv'])
plt.grid()

```
